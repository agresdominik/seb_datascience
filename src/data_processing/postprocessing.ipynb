```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from scipy.optimize import curve_fit

def exponential_decay(t, delta_T0, k):
    return delta_T0 * np.exp(-k * t)

data_results = pd.read_csv('/project/seb_datascience/src/data_processing/analasys_output_2024-12-17_10-42-44.csv', parse_dates=['start_time'])

# Step 1: Fit an exponential model to find k
time_data = np.array([0, 5, 10, 15, 20])  # Example time points
temp_diff_data = np.array([30, 25, 20, 15, 10])  # Example temperature differences

# Fit the exponential decay curve
params, _ = curve_fit(exponential_decay, time_data, temp_diff_data)
delta_T0, k = params
print(f"Fitted initial temperature difference (delta_T0): {delta_T0}")
print(f"Fitted cooling constant (k): {k}")

# Step 2: Predict temperature difference dynamically
def predict_temperature_over_time(delta_T0, k, total_minutes, step=1):
    temperatures = []
    time_points = np.arange(0, total_minutes + step, step)
    for t in time_points:
        temp_diff = delta_T0 * np.exp(-k * t)
        temperatures.append(temp_diff)
    return time_points, temperatures

# Predict over 30 minutes
total_minutes = 30
time_points, predicted_temps = predict_temperature_over_time(delta_T0, k, total_minutes)

# Print results
for t, temp in zip(time_points, predicted_temps):
    print(f"At time {t} minutes, temperature difference: {temp:.2f} degrees")
```
